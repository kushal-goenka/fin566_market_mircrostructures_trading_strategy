```python
import pandas as pd
from pandas import DataFrame
import numpy as np
import glob
import os
os.chdir('/Users/khavya/Box/MS - Financial Engineering/Fall 2020/FIN566 - Algorithmic Market Microstructure')
import matplotlib.pyplot as plt
```

```python
# Setting the path where the files to be loaded are located
all_files = glob.glob('ticks_201910/tick_*.txt')

# Loading the all the files associated with the ticker SPY
spy = []
for filename in all_files:
    df = pd.read_csv(filename, index_col=None, header=0)
    df.columns = ['Collection_time', 'source_time', 'seq_num', 'tick_type', 'market_center', 'price', 'size']
    spy.append(df)

# Concatenating all the files that were loaded into a single file
spy_data = pd.concat(spy, axis=0, ignore_index=True)
spy_data['ticker'] = 'SPY'
print('The number of trades that happened in SPY for October 2020 is: ' + str(len(spy_data)))

# Loading the five highest weighted stocks from SPX 
# Source for finding most weighted stocks: 
# https://www.investopedia.com/ask/answers/08/find-stocks-in-sp500.asp

# AAPL - Apple Inc.
all_files = glob.glob('ticks_201910/tick_AAPL*.txt')
aapl = []
for filename in all_files:
    df = pd.read_csv(filename, index_col=None, header=0)
    df.columns = ['Collection_time', 'source_time', 'seq_num', 'tick_type', 'market_center', 'price', 'size']
    aapl.append(df)
aapl_data = pd.concat(aapl, axis=0, ignore_index=True)
aapl_data['ticker'] = 'AAPL'
print('The number of trades that happened in AAPL for October 2020 is: ' + str(len(aapl_data)))

# MSFT - Microsoft Corporation
all_files = glob.glob('ticks_201910/tick_MSFT*.txt')
msft = []
for filename in all_files:
    df = pd.read_csv(filename, index_col=None, header=0)
    df.columns = ['Collection_time', 'source_time', 'seq_num', 'tick_type', 'market_center', 'price', 'size']
    msft.append(df)
msft_data = pd.concat(msft, axis=0, ignore_index=True)
msft_data['ticker'] = 'MSFT'
print('The number of trades that happened in MSFT for October 2020 is: ' + str(len(msft_data)))

# AMZN - Amazon.com Inc.
all_files = glob.glob('ticks_201910/tick_AMZN*.txt')
amzn = []
for filename in all_files:
    df = pd.read_csv(filename, index_col=None, header=0)
    df.columns = ['Collection_time', 'source_time', 'seq_num', 'tick_type', 'market_center', 'price', 'size']
    amzn.append(df)
amzn_data = pd.concat(amzn, axis=0, ignore_index=True)
amzn_data['ticker'] = 'AMZN'
print('The number of trades that happened in AMZN for October 2020 is: ' + str(len(amzn_data)))

# FB - Facebook Inc.
all_files = glob.glob('ticks_201910/tick_FB*.txt')
fb = []
for filename in all_files:
    df = pd.read_csv(filename, index_col=None, header=0)
    df.columns = ['Collection_time', 'source_time', 'seq_num', 'tick_type', 'market_center', 'price', 'size']
    fb.append(df)
fb_data = pd.concat(fb, axis=0, ignore_index=True)
fb_data['ticker'] = 'FB'
print('The number of trades that happened in FB for October 2020 is: ' + str(len(fb_data)))

# GOOG - Google 
all_files = glob.glob('ticks_201910/tick_GOOG*.txt')
goog = []
for filename in all_files:
    df = pd.read_csv(filename, index_col=None, header=0)
    df.columns = ['Collection_time', 'source_time', 'seq_num', 'tick_type', 'market_center', 'price', 'size']
    goog.append(df)
goog_data = pd.concat(goog, axis=0, ignore_index=True)
goog_data['ticker'] = 'GOOG'
print('The number of trades that happened in GOOG for October 2020 is: ' + str(len(goog_data)))
```

```python
# We can do a time series plot to see if SPY changes
# in accordance to it's 5 most weighted components
y1 = spy_data['price']
y2 = aapl_data['price']
y3 = msft_data['price']
y4 = amzn_data['price']
y5 = fb_data['price']
y6 = goog_data['price']

#Converting the above data into series for plotting 
def createSeries (series_list): 
    series_list = pd.Series(series_list)  
    return series_list 

y1 = createSeries(y1)
y2 = createSeries(y2)
y3 = createSeries(y3)
y4 = createSeries(y4)
y5 = createSeries(y5)
y6 = createSeries(y6)

data = {"SPY": y1, "AAPL": y2, "MSFT": y3, "AMZN": y4, "FB": y5, "GOOG": y6} 
m = pd.concat(data, axis = 1) 
```

```python
# Plotting the data for SPY
#plt.figure()
positions = (0, 25000, 50000, 75000, 100000, 125000, 150000)
labels = ('10/01', '10/07', '10/11', '10/17', '10/21', '10/27', '10/31')
fig, ax = plt.subplots()
plt.gca().yaxis.grid(True)
plt.xticks(positions, labels, rotation='90')
ax.set_xlabel('Date')
ax.set_ylabel('Price')
ax.set_title('Time series plot of SP 500 ETF')
ax.plot(m['SPY'])
ax.spines['right'].set_visible(False)
ax.spines['top'].set_visible(False)
ax.spines['bottom'].set_visible(False)
ax.spines['left'].set_visible(False)
#plt.savefig('SPY.png', dpi = 400)
plt.show()
```
